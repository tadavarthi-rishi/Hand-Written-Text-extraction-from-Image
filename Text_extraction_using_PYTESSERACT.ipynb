```
a='/content/004.jpg'
```

```
import cv2
import numpy as np
import pytesseract
from PIL import Image
import matplotlib.pyplot as plt

```

```
img = cv2.imread(a)
plt.imshow(img)
plt.show()
```

```
def getText(img_path):
  img=cv2.imread(img_path)
  img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
  mask=np.ones((1,1),np.uint8)
  img=cv2.dilate(img,mask,iterations=1)
  img=cv2.erode(img,mask,iterations=1)
  cv2.imwrite("removed_noise.png",img)
  cv2.imwrite(img_path,img)
  text=pytesseract.image_to_string(Image.open(img_path))
  return text
print(" the recognition process of text from input image has started successfully")
print("the recognised text from input image is")
print(getText(a))
print("process finished")
```
